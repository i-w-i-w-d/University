### 1. Властивості циклічних груп

Для $n=100,173,530,746,911,1024$ розглянемо циклічну групу $\mathbb{Z}_n$ (або $C_n$):
1. Для кожного дільника $d$ порядку $n$:<br>
   1.1 Знайдіть кількість елементів порядку $d$ в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   1.2 Порівняйте з $\phi(d)$.
2. Знайдіть кількість підгруп в групі $\mathbb{Z}_n$ (або $C_n$).<br>
   Порівняйте з $\tau(n)$.

Проаналізуйте результати та дайте відповіді на запитання:
1. Скільки елементів порядку $k$ в циклічній групі порядку $n$?
2. Cкільки підгруп в циклічній групі порядку $n$?

```python
numbers = [100, 173, 530, 746, 911, 1024]

for n in numbers:
    print(f"\n=== n = {n} ===")
    
    divisors_n = divisors(n)
    print(f"Дільники n: {divisors_n}")
    
    tau_n = len(divisors_n)
    print(f"Кількість підгруп (n): {tau_n}")
    
    for d in divisors_n:
        phi_d = euler_phi(d)
        print(f"d = {d}: елементів порядку d = ({d}) = {phi_d}")
    
    print(f"Перевірка: сума (d) по всіх d|n = {sum(euler_phi(d) for d in divisors_n)} (має бути {n})")
```

### 2. Порядки елементів в дієдральних групах

Для $n=3,4,...,20$ розглянемо дієдральну групу $D_n$:   
1. Знайдіть кількість елементів порядку $d=2$ в групі $D_n$.
2. Для кожного дільника $d>2$ порядку $|D_n|=2n$:<br>
   Знайдіть кількість елементів порядку $d$ в групі $D_n$.

Проаналізуйте результати та дайте відповіді на запитання:
1. Скільки елементів порядку $d=2$ в дієдральній групі $D_n$?
2. Скільки елементів порядку $d$ в дієдральній групі $D_n$?
3. Чи для кожного дільника $d<2n$ порядку групи $D_n$ існує елемент порядку $d$?

```python
def count_elements_of_order_d(D, d):
    count = 0
    for g in D:
        if g.order() == d:
            count += 1
    return count

for n in range(3, 21):
    D = DihedralGroup(n)
    print(f"\n--- D_{n} (порядок {2*n}) ---")
    
    ord2 = count_elements_of_order_d(D, 2)
    print(f"Елементи порядку 2: {ord2}")
    
    possible_orders = set(d for d in range(3, 2*n+1) if (2*n) % d == 0)
    for d in sorted(possible_orders):
        c = count_elements_of_order_d(D, d)
        if c > 0:
            print(f"Елементи порядку {d}: {c}")
```

### 3. Дискретний логарифм в групі $\mathbb{Z}_n^{*}$

Розглянемо групу $\mathbb{Z}_n^{*}$ для $n=1102158078129785185997827239806$.
1. Перевірте, що група є циклічною і знайдіть її порядок.
2. Знайдіть найменше натуральне $a\in\mathbb{N}$ таке, що елемент $\overline{a}\in\mathbb{Z}_n^{*}$ є твірним групи.
3. Перевірте, що порядок елемента $\overline{a}$ дорівнює $\phi(n)$.
4. Виберіть три випадкових елементи $g_1,g_2,g_3$ групи $\mathbb{Z}_n^{*}$.
5. Для кожного $g$ знайдіть степінь $k$ такий, що $g=\overline{a}^k$.


```python
n = 1102158078129785185997827239806

print("Аналіз групи для складеного n")
print("==================================")

print(f"n = {n}")

print("\n1. Розклад n на множники:")
factors = factor(n)
print(f"n = {factors}")

order = euler_phi(n)
print(f"\nПорядок групи = (n) = {order}")

print(f"\n2. Циклічність групи:")
print("Для складеного n група циклічна тільки якщо:")
print("n = 2, 4, p^k або 2p^k, де p - непарне просте число")

if n == 2 or n == 4:
    print("Група циклічна (n = 2 або 4)")
elif len(factors) == 1:
    p, k = factors[0]
    if p == 2 and k >= 3:
        print("Група не циклічна (n = 2^k, k ≥ 3)")
    elif p % 2 == 1:
        print("Група циклічна (n = p^k, p - непарне просте)")
    else:
        print("Група не циклічна")
elif len(factors) == 2 and factors[0] == (2, 1): 
    p, k = factors[1]
    if p % 2 == 1:
        print("Група циклічна (n = 2 * p^k, p - непарне просте)")
    else:
        print("Група не циклічна")
else:
    print("Група не циклічна")

print("\n3. Знаходження максимального порядку елемента")
print("============================================")

Z = Integers(n)
max_order = 0
best_a = None

for test_a in range(2, min(10000, n)):
    if gcd(test_a, n) == 1:
        element_order = Z(test_a).multiplicative_order()
        if element_order > max_order:
            max_order = element_order
            best_a = test_a
            if max_order == order:
                break

print(f"Елемент з максимальним порядком: a = {best_a}")
print(f"Його порядок: {max_order}")
print(f"Повний порядок групи: {order}")
print(f"Елемент є твірним: {max_order == order}")

print("\n4. Вибір випадкових елементів групи")
print("===================================")

random_elements = []
for i in range(3):
    while True:
        g = randint(2, n-1)
        if gcd(g, n) == 1:
            random_elements.append(g)
            break

print("Випадкові елементи групи:")
for i, g in enumerate(random_elements, 1):
    print(f"g_{i} = {g}")

print("\n5. Спроба розв'язання задачі дискретного логарифму")
print("=================================================")

if best_a is not None:
    a = best_a
    print(f"Використовуємо елемент a = {a} з порядком {max_order}")
    
    for i, g in enumerate(random_elements, 1):
        print(f"\nДля g_{i} = {g}:")
        
        g_order = Z(g).multiplicative_order()
        
        if max_order % g_order != 0:
            print(f"  g_{i} не належить підгрупі, породженій a")
            continue
            
        try:
            k = discrete_log(Z(g), Z(a))
            print(f"  Розв'язок: {a}^{k} = {g} (mod {n})")
            print(f"  Перевірка: {pow(a, k, n)} = {g}")
            
            period = max_order
            print(f"  Всі розв'язки: k = {k} (mod {period})")
            
        except ValueError as e:
            print(f"  Не вдалося знайти дискретний логарифм: {e}")
            print(f"  Можливо, g_{i} не належить підгрупі, породженій a")
else:
    print("Не знайдено підходящого елемента a")

print("\n6. Додаткова інформація про структуру групи")
print("===========================================")

if len(factors) > 1:
    print("Структура групи за китайською теоремою про залишки:")
    for p, exp in factors:
        p_power = p^exp
        if gcd(p_power, n // p_power) == 1:
            phi_p = euler_phi(p_power)
            print(f"  Z_{{{p_power}}}^* = Z_{{{phi_p}}}{' (циклічна)' if p % 2 == 1 or p_power == 2 or p_power == 4 else ''}")
```
